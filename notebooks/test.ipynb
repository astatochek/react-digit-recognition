```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from torchvision import transforms
from torchvision.transforms import ToTensor
%matplotlib inline
```

```python
img = Image.open("num.png")
plt.imshow(img)
```

```python
raw = (transforms.Compose([transforms.ToTensor()]))(img)
raw = raw[0]
raw = raw.unsqueeze(0)
raw[0]
```

```python
compressed = []
for i in range(32):
    N = []
    for j in range(32):
        a = 0
        for k in range(6*i, 6*i+ 6):
            for l in range(6*j, 6*j + 6):
                a += raw[0][k][l]
        a /= 36
        N.append(a)
    compressed.append(N)
compressed = np.array(compressed)
plt.imshow(compressed)
plt.show()
```
